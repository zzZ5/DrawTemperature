```python
import pandas
import numpy as np
import datetime
```

```python
path_pre = 'data_o2-2'

for i in range(1, 6):
    data = pandas.read_csv(path_pre + '/' + 'CO2-' + str(i) + '.csv')
    data['Time'] = data['Time'].astype('datetime64[ns]')
    # data['O2'] = (data['Value'] - 400) * (i *0.2 - 0.1) * 1.429 / 1000000
    data['O2'] = (data['Value'] - 400) * 0.3 / 1000000
    data['newindex'] = np.arange(len(data)-1, -1, -1)
    data.sort_values('newindex', inplace=True)
    data.drop('newindex', axis=1, inplace=True)
    # Initialize a dictionary to accumulate the oxygen consumption for each day
    daily_oxygen_consumption = {}
    
    index = 0
    last_time = ''
    # Loop through the data, group by date, and accumulate the oxygen consumption for each day
    two_hours = datetime.timedelta(hours=2)
    ten_hours = datetime.timedelta(hours=12)

    is_dan = 1
    for item in data.values:
        # if is_dan == 0:
        #     is_dan = 1
        #     continue
        # if is_dan == 1:
        #     is_dan = 0
        # if i == 3:
        #     timestamp = timestamp + ten_hours

        timestamp = item[0] + two_hours
        if index == 0:
            # Convert the timestamp string to a datetime object
            # Get the date portion
            date = timestamp.date()
            # If this date is not in the dictionary yet, add it with 0 oxygen consumption
            if date not in daily_oxygen_consumption:
                daily_oxygen_consumption[date] = 0
            # Add the oxygen consumption for this data point to the total for this date
            daily_oxygen_consumption[date] += item[2]
        else:
            # Get the date portion
            date = timestamp.date()
            # If this date is not in the dictionary yet, add it with 0 oxygen consumption
            if date not in daily_oxygen_consumption:
                daily_oxygen_consumption[date] = 0
            # Add the oxygen consumption for this data point to the total for this date
            minites = int((timestamp - last_time).total_seconds() / 60)
            daily_oxygen_consumption[date] += item[2] * minites

        last_time = timestamp
        index += 1


    # Print the daily oxygen consumption
    print('data{}'.format(i))
    for date, consumption in daily_oxygen_consumption.items():
        print(f"{date}: {consumption} mg")

```
